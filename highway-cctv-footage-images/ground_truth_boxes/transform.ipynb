```python
import json
from pprint import pprint

with open('labelbox.json') as data_file:    
    data = json.load(data_file)
```

```python
f = open('labelbox.txt','w+')
for datum in data:
#     pprint(datum['Label']['Car'][0]['geometry'])
#     pprint(datum['External ID'])
    k = ''
    for datum2 in datum['Label']['Car']:
#         pprint(datum2['geometry'])
        x, y = [], []
        
        for datum3 in datum2['geometry']:
#             pprint(datum3)
            x.append(datum3['x'])
            y.append(datum3['y'])
#         print("{:d},{:d},{:d},{:d},0".format(min(x),min(y),max(x),max(y)))
        k = k + ' ' + "{:d},{:d},{:d},{:d},0".format(min(x),min(y),max(x),max(y))
    k = 'highway-cctv-footage-images/input_images/' + datum['External ID'] + k
    print(k)
    f.write(k + '\n')
    
f.close()
```
